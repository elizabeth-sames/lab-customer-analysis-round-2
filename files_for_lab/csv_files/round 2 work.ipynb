```python
import pandas as pd
mca = pd.read_csv('marketing_customer_analysis.csv')
mca = mca.drop(['Unnamed: 0'], axis =1)
mca
```

```python
#1 Show the dataframe shape.
mca.shape
```

```python
#2 Standardize header names.
cols = []
for column in mca.columns:
    cols.append(column.lower())
mca.columns = cols
mca.head()
```

```python
#3 Which columns are numerical?
mca.select_dtypes('number')
```

```python
#4 Which columns are categorical?
mca.select_dtypes('object')
```

```python
import numpy as np
only_months = mca[mca['months since last claim'].isna()==False]

mean_months = np.mean(only_months['months since last claim'])
median_months = np.median(only_months['months since last claim'])
print('mean', mean_months, ', median', median_months)
#(mca['months since last claim'].value_counts(dropna=False))/10910 * 100
```

```python
only_complaints = mca[mca['number of open complaints'].isna()==False]

mean_complaints = np.mean(only_complaints['number of open complaints'])
median_complaints = np.median(only_complaints['number of open complaints'])
print('mean', mean_complaints, ', median', median_complaints)
(mca['number of open complaints'].value_counts(dropna=False))/10910 * 100
```
