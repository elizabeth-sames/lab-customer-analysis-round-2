```python
import pandas as pd
mca = pd.read_csv('marketing_customer_analysis.csv')
mca = mca.drop(['Unnamed: 0'], axis =1)
mca
```

```python
#1 Show the dataframe shape.
mca.shape
```

```python
#2 Standardize header names.
cols = []
for column in mca.columns:
    cols.append(column.lower())
mca.columns = cols
mca.head()
```

```python
#3 Which columns are numerical?
mca.select_dtypes('number')
```

```python
#4 Which columns are categorical?
mca.select_dtypes('object')
```

```python
#5 Check and deal with NaN values.
#mca.isna().sum() shows total number of null values in each column
round(mca.isna().sum()/len(mca),4)*100  # shows percent of null values in column
```

```python
#(mca['state'].value_counts(dropna=False))/10910 * 100
#state:changed to mode (california 32%)
mca['state']=mca['state'].fillna('California')
```

```python
#reponse: NaN changed to 'No' (mode at 80%)
mca['response']=mca['response'].fillna('No')
```

```python
#(mca['vehicle class'].value_counts(dropna=False))/10910 * 100
#vehicle class: change to mode (4 door 48%)
mca['vehicle class']=mca['vehicle class'].fillna('Four-Door Car')
```

```python
#(mca['vehicle size'].value_counts(dropna=False))/10910 * 100
#vehicle size: change to mode (medsize 66%)
mca['vehicle size']=mca['vehicle size'].fillna('Medsize')
```

```python
import numpy as np
#only_months = mca[mca['months since last claim'].isna()==False]

#mean_months = np.mean(only_months['months since last claim'])
#median_months = np.median(only_months['months since last claim'])
#print('mean', mean_months, ', median', median_months)
#(mca['months since last claim'].value_counts(dropna=False))/10910 * 100
#mean is 15, median is 14

#months since last claim: changed to mean (15)
mca['months since last claim']=mca['response'].fillna(15)
```

```python
#only_complaints = mca[mca['number of open complaints'].isna()==False]

#mean_complaints = np.mean(only_complaints['number of open complaints'])
#median_complaints = np.median(only_complaints['number of open complaints'])
#print('mean', mean_complaints, ', median', median_complaints)
#(mca['number of open complaints'].value_counts(dropna=False))/10910 * 100

#number of open complaints: changed to 0.0 (mean/median/mode all 0.0)
mca['number of open complaints']=mca['number of open complaints'].fillna(0.0)
```

```python
#(mca['vehicle type'].value_counts(dropna=False))/10910 * 100
#vehicle type has a high rate of NaN values at 50% and the 
#only other value in this column is 'A', so it seems better 
#to me to drop the entire column instead of deleting all of 
#the rows with Nan values in this column and losing all that data.
mca = mca.drop(['vehicle type'], axis=1)
mca
```

```python
#6 Datetime format - Extract the months from the dataset and store in a separate column. 
mca2 = mca.copy()
mca2['effective to date'] = pd.to_datetime(mca2['effective to date'], errors='coerce')
#mca2['effective to date'][0].month
def month_to_month(x):
    months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
    m = x.month
    y = months[m-1]
    return y
    

mca2['month'] = list(map(month_to_month, mca2['effective to date']))
    
mca2

    
#mca2['month'] = list(map(month_to_month, mca2['effective to date']))
    
#mca2
```

```python
#Show a plot of the total number of responses.
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.displot(mca['response'], bins=2)
plt.show()
```

```python
#Show a plot of the response rate by the sales channel.
sns.countplot(x='sales channel', hue='response', data = mca)
plt.show()
```

```python

def resp_rate(x):
    if x=='No':
        return 0
    else:
        return 1
    
mca['response rate'] = list(map(resp_rate, mca['response']))
sns.barplot(x='sales channel',y='response rate', data=mca )
plt.show()
```

```python
#Show a plot of the response rate by the total claim amount.
#sns.barplot(x='response', y='total claim amount', data = mca)
#sns.scatterplot(x='response', y='total claim amount', data = mca)
#sns.barplot(x='total claim amount',y='response rate', data=mca)
sns.catplot(x="response",y="total claim amount", data=mca,jitter="0.45")
plt.show()
```

```python
#Show a plot of the response rate by income.
#sns.barplot(x='income', y='response rate', data = mca)
sns.catplot(x="response",y="income", data=mca,jitter="0.45")
plt.show()
```
